## 2. Entropi Dataset Awal (Node Induk)

Terdapat total 15 data, yang terdiri dari 6 data dengan kelas A dan 9 data dengan kelas B. Probabilitas kemunculan masing-masing kelas dapat dihitung sebagai berikut:
$$
\begin{gathered}
p_A=\frac{6}{15}=0,4, \quad p_B=\frac{9}{15}=0,6 \\
\text { Entropi }(S)=-\left(p_A \cdot \log _2 p_A+p_B \cdot \log _2 p_B\right) \\
=-\left(0,4 \cdot \log _2 0,4+0,6 \cdot \log _2 0,6\right) \\
=-(0,4 \cdot(-1,322)+0,6 \cdot(-0,737)) \\
=0,529+0,442=0,971
\end{gathered}
$$

Dengan demikian, nilai entropi dari node induk (dataset awal) adalah 0,971, yang mencerminkan tingkat ketidakpastian atau impuritas data sebelum dilakukan pemisahan berdasarkan atribut tertentu.

### Entropi (Subset Kiri)
Data pada subset kiri merupakan hasil pemisahan berdasarkan kondisi Panjang $\leq 235$, dengan total 9 data, erdiri dari 1 data kelas A dan 8 data kelas B. Nilai entropi untuk subset ini dihitung sebagai berikut:
$$
\begin{gathered}
\quad p_A=\frac{1}{9} \approx 0,111, \quad p_B=\frac{8}{9} \approx 0,889 \\
\text { Entropi }(\text { Left })=-\left(p_A \cdot \log _2 p_A+p_B \cdot \log _2 p_B\right) \\
=-\left(0,111 \cdot \log _2 0,111+0,889 \cdot \log _2 0,889\right) \\
=-(0,111 \cdot(-3,17)+0,889 \cdot(-0,17)) \\
=0,352+0,151=0,503
\end{gathered}
$$

Dengan demikian, nilai entropi dari subset kiri adalah sekitar 0,503, yang mengindikasikan impuritas relatif endah karena dominasi kelas B dalam kelompok tersebut.

###  Entropi Subset Kanan (*Panjang > 235*)

Terdapat 6 data pada subset ini, terdiri dari 5 data kelas A dan 1 data kelas B. Maka, probabilitas masing-masing kelas adalah:

$$
p_A = \frac{5}{6} \approx 0{,}833,\quad p_B = \frac{1}{6} \approx 0{,}167
$$

Nilai entropi untuk subset kanan dihitung sebagai berikut:

$$
\begin{gathered}
\operatorname{Entropy}(\text{Right}) = -\left(p_A \cdot \log_2 p_A + p_B \cdot \log_2 p_B\right) \\
= -\left(0{,}833 \cdot \log_2 0{,}833 + 0{,}167 \cdot \log_2 0{,}167\right) \\
= -\left(0{,}833 \cdot (-0{,}263) + 0{,}167 \cdot (-2{,}585)\right) \\
= 0{,}219 + 0{,}431 = \mathbf{0{,}650}
\end{gathered}
$$

###  Information Gain Berdasarkan Atribut *Panjang*

Dengan menggunakan nilai entropi total $\text{Entropy}(S) = 0{,}971$, serta entropi masing-masing subset:

* Entropy(Left) = 0,503
* Entropy(Right) = 0,650

Maka, nilai **Information Gain (IG)** dihitung sebagai:

$$
\begin{gathered}
IG = \operatorname{Entropy}(S) - \left( \frac{9}{15} \cdot \operatorname{Entropy}(\text{Left}) + \frac{6}{15} \cdot \operatorname{Entropy}(\text{Right}) \right) \\
= 0{,}971 - \left( \frac{9}{15} \cdot 0{,}503 + \frac{6}{15} \cdot 0{,}650 \right) \\
= 0{,}971 - (0{,}302 + 0{,}260) = 0{,}971 - 0{,}562 = \mathbf{0{,}409}
\end{gathered}
$$

## Visualisasi Decision Tree

Kode dibawah merupakan implementasi sederhana dari algoritma *Decision Tree* (pohon keputusan) menggunakan pustaka `scikit-learn`. Data yang digunakan terdiri dari atribut *Panjang (cm)* dan *Berat (kg)* untuk mengklasifikasikan objek ke dalam dua kelas, yaitu A dan B. Label kelas yang semula berupa huruf dikonversi menjadi angka (A → 0, B → 1) agar dapat diproses oleh model. Model pohon keputusan dibangun dengan menggunakan *entropy* sebagai kriteria pemisahan (dalam hal ini, menggunakan algoritma ID3) dan dibatasi kedalamannya hingga maksimal 3 level (agar pohon tidak terlalu kompleks).

Setelah model dilatih, aturan-aturan keputusan dari pohon ditampilkan dalam bentuk teks, yang menunjukkan kondisi cabang berdasarkan nilai atribut *Panjang* dan *Berat*. Selain itu, visualisasi grafis pohon juga ditampilkan menggunakan `matplotlib`, yang memperlihatkan bagaimana data dipisahkan di setiap node berdasarkan nilai fitur dengan mempertimbangkan *Information Gain* tertinggi. Warna pada node menunjukkan kelas dominan dan tingkat keyakinan klasifikasi, memudahkan interpretasi keputusan yang diambil model terhadap input baru.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text
from sklearn.preprocessing import LabelEncoder

# === Data Baru (Sesuai yang Kamu Berikan) ===
data = {
    "Panjang (cm)": [205, 260, 255, 215, 250, 222, 242, 210, 225, 240, 218, 233, 212, 257, 219],
    "Berat (kg)":   [198, 240, 235, 210, 245, 212, 238, 200, 214, 230, 207, 220, 203, 250, 208],
    "Kelas":        ['B', 'A', 'A', 'B', 'A', 'B', 'A', 'B', 'B', 'A', 'B', 'A', 'B', 'A', 'B']
}
df = pd.DataFrame(data)

# === Konversi label kelas dari kategorikal ('A', 'B') ke numerik (0, 1) ===
le = LabelEncoder()
df['Label'] = le.fit_transform(df['Kelas'])  # A -> 0, B -> 1

# === Pemisahan fitur dan target ===
X = df[["Panjang (cm)", "Berat (kg)"]]
y = df["Label"]

# === Pelatihan model pohon keputusan menggunakan entropi sebagai kriteria ===
clf = DecisionTreeClassifier(criterion="entropy", max_depth=3, random_state=42)
clf.fit(X, y)

# === Menampilkan aturan keputusan dalam bentuk teks ===
print("\n=== Aturan Pohon Keputusan ===")
tree_rules = export_text(clf, feature_names=list(X.columns))
print(tree_rules)

# === Visualisasi struktur pohon keputusan ===
plt.figure(figsize=(12, 6))
plot_tree(clf,
          feature_names=["Panjang (cm)", "Berat (kg)"],
          class_names=le.classes_,
          filled=True,
          rounded=True,
          fontsize=10)
plt.title("Visualisasi Pohon Keputusan Berdasarkan Data Panjang dan Berat")
plt.show()
```

## Visualisasi Splitting Setelah ada 2 Data Baru

Kode di bawah merupakan lanjutan dari implementasi *Decision Tree Classifier* untuk mengklasifikasikan data berdasarkan atribut panjang dan berat tubuh. Setelah model dilatih menggunakan data pengamatan yang telah diberi label kelas A atau B, dilakukan prediksi terhadap dua data baru dengan panjang 242 cm dan 213 cm serta berat 229 kg dan 207 kg. Hasil prediksi menunjukkan kelas yang sesuai berdasarkan aturan yang telah dipelajari oleh model dari data sebelumnya.

Selain itu, kode ini juga menyertakan visualisasi *decision boundary*, yaitu batas wilayah keputusan model pada ruang dua dimensi (panjang vs berat). Area berwarna mewakili keputusan model terhadap kelas A atau B, sedangkan titik-titik menunjukkan data pelatihan dan dua data baru (berwarna kuning). Visualisasi ini sangat membantu untuk memahami bagaimana model memisahkan kelas dan di mana posisi data baru relatif terhadap wilayah klasifikasi yang telah ditentukan. Prediksi untuk kedua data baru ditampilkan langsung di grafik dan juga dicetak ke layar.


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder

# === Data Baru dari Pengamatan ===
data = {
    "Panjang (cm)": [205, 260, 255, 215, 250, 222, 242, 210, 225, 240, 218, 233, 212, 257, 219],
    "Berat (kg)":   [198, 240, 235, 210, 245, 212, 238, 200, 214, 230, 207, 220, 203, 250, 208],
    "Kelas":        ['B', 'A', 'A', 'B', 'A', 'B', 'A', 'B', 'B', 'A', 'B', 'A', 'B', 'A', 'B']
}
df = pd.DataFrame(data)

# === Konversi Label Kelas ===
le = LabelEncoder()
df['Label'] = le.fit_transform(df['Kelas'])  # A = 0, B = 1

# === Pisahkan Fitur dan Target ===
X = df[["Panjang (cm)", "Berat (kg)"]]
y = df["Label"]

# === Training Model Pohon Keputusan ===
clf = DecisionTreeClassifier(criterion="entropy", max_depth=3, random_state=42)
clf.fit(X, y)

# === Prediksi Dua Data Baru ===
new_data = pd.DataFrame({
    "Panjang (cm)": [242, 213],
    "Berat (kg)":   [229, 207]
})
preds = clf.predict(new_data)
predicted_labels = le.inverse_transform(preds)

# === Visualisasi Decision Boundary ===
x_min, x_max = X["Panjang (cm)"].min() - 5, X["Panjang (cm)"].max() + 5
y_min, y_max = X["Berat (kg)"].min() - 5, X["Berat (kg)"].max() + 5
xx, yy = np.meshgrid(np.arange(x_min, x_max, 1),
                     np.arange(y_min, y_max, 1))

Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)

# Plot data latih
plt.scatter(X["Panjang (cm)"], X["Berat (kg)"], c=y, cmap=plt.cm.coolwarm, edgecolor="k", s=60, label="Data Latih")

# Plot data baru
plt.scatter(new_data["Panjang (cm)"], new_data["Berat (kg)"], c="yellow", edgecolor="black", s=100, label="Data Baru")
for i, label in enumerate(predicted_labels):
    plt.text(new_data.iloc[i, 0] + 0.5, new_data.iloc[i, 1], label, fontsize=10, weight='bold')

plt.xlabel("Panjang (cm)")
plt.ylabel("Berat (kg)")
plt.title("Decision Boundary dengan Data Baru")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# === Cetak hasil prediksi ===
print("Hasil prediksi untuk data baru:")
for i in range(len(new_data)):
    print(f"Data ke-{i+1}: Panjang = {new_data.iloc[i,0]}, Berat = {new_data.iloc[i,1]} → Prediksi Kelas = {predicted_labels[i]}")
```
